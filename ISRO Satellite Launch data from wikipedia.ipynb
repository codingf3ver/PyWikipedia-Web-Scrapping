```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
#Taking the desired URL
url=requests.get('https://en.wikipedia.org/wiki/List_of_Indian_satellites').text #Here 'text' is only to get tags in HTML

#Creating a Beautiful soup object for parsing XML or HTML tags
soup=BeautifulSoup(url,'lxml')
```

```python
#Chosing the desired table that you want to dig out
wikitable=soup.find('table',class_='wikitable plainrowheaders')
wikitable
```

```python
raw_data=[]

#finding the data in a table with howering the cursor over tag in developer mode of URL
for i in wikitable.tbody.find_all('tr'): 
    j=i.text.split('\n')    
    raw_data.append(list(j))
raw_data 
```

```python
#creating raw dataframe by passing list as an argument
df_unstruct=pd.DataFrame(raw_data) 
df_unstruct
```

```python
#Setting first row as Header 
df_header=df_unstruct.iloc[0]

#seleting table data other than first row
df_unstruct=df_unstruct[1:]

#setting column names as above 
df_unstruct.columns=df_header
df_unstruct
```

```python
#converting into 'CSV File' so that we can wrangle the data
df_unstruct.to_csv('ISRO Satellite Launch Data.csv')
```
